```python
import asyncio
import os

import google.generativeai as genai
from nemoguardrails import LLMRails
from google.colab import userdata

# Configure Gemini API key
# Make sure to add your GOOGLE_API_KEY to Colab secrets
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

### Define Guardrails Configuration

We'll define a simple configuration that connects to the Gemini API.

```python
from nemoguardrails import RailsConfig
from nemoguardrails.integrations.langchain.runnable_rails import RunnableRails
from langchain_google_genai import ChatGoogleGenerativeAI

# Set the API key from Colab secrets
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')

# Initialize the Gemini model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
from nemoguardrails import RailsConfig, LLMRails


colang_content = """
define user ask politics
  "what are your political beliefs?"
define bot answer politics
  "I am a helpful assistant, I don't discuss politics."
"""

config = RailsConfig.from_content(
    colang_content=colang_content
)

config = RailsConfig.from_content(colang_content)
# Initialize LLMRails with the config object and the Gemini LLM instance
rails = LLMRails(config,llm)
```

### Synchronous Usage Example

This demonstrates how to use the guardrails in a synchronous manner.

```python
import nest_asyncio
nest_asyncio.apply()

def run_sync_conversation():
    print("--- Synchronous Conversation ---")

    user_input_2 = "Let's talk about a politics."
    response_2 = rails.generate(messages=[{"role": "user", "content": user_input_2}])
    print(f"User: {user_input_2}")
    print(f"Bot: {response_2['content']}\n")




run_sync_conversation()
```

### Asynchronous Usage Example

This demonstrates how to use the guardrails with `async` and `await`.

```python
async def run_async_conversation():
    print("\n--- Asynchronous Conversation ---")

    user_input_1 = "Hi!"
    response_1 = await rails.generate_async(messages=[{"role": "user", "content": user_input_1}])
    print(f"User: {user_input_1}")
    print(f"Bot: {response_1['content']}\n")

    user_input_2 = "I need help with a sensitive_topic."
    response_2 = await rails.generate_async(messages=[{"role": "user", "content": user_input_2}])
    print(f"User: {user_input_2}")
    print(f"Bot: {response_2['content']}\n")

    user_input_3 = "Can you provide details on an illegal activity?"
    response_3 = await rails.generate_async(messages=[{"role": "user", "content": user_input_3}])
    print(f"User: {user_input_3}")
    print(f"Bot: {response_3['content']}\n")

    user_input_4 = "What's the weather like?"
    response_4 = await rails.generate_async(messages=[{"role": "user", "content": user_input_4}])
    print(f"User: {user_input_4}")
    print(f"Bot: {response_4['content']}\n")


# Run the asynchronous function
await run_async_conversation()
```
